```python
select_database_query = "USE WHO_Vaccination"

with connection.cursor() as cursor:
    cursor.execute(select_database_query)

Distinct_Columns_query = """
SELECT DISTINCT `group`,Code,Name FROM coverage 
"""

with connection.cursor() as cursor:
    cursor.execute(Distinct_Columns_query)
    result = cursor.fetchall()
    for row in result:
        print(row)
```

```python
select_database_query = "USE WHO_Vaccination"

with connection.cursor() as cursor:
    cursor.execute(select_database_query)

create_new_table = """
CREATE Table regioncode As SELECT DISTINCT `group`,Code,Name FROM coverage 
"""

with connection.cursor() as cursor:
    cursor.execute(create_new_table)
```

```python
select_database_query = "USE WHO_Vaccination"

with connection.cursor() as cursor:
    cursor.execute(select_database_query)

create_new_table = """
CREATE Table diseases SELECT DISTINCT Disease,Disease_Description from incidence
"""

with connection.cursor() as cursor:
    cursor.execute(create_new_table)
```

```python
select_database_query = "USE WHO_Vaccination"

with connection.cursor() as cursor:
    cursor.execute(select_database_query)

create_new_table = """
CREATE Table WHO_region SELECT DISTINCT ISO_3_CODE,COUNTRYNAME,WHO_REGION from vacintro
"""

with connection.cursor() as cursor:
    cursor.execute(create_new_table)
```
